filtering:

```python
high_sales_Amount = df["Sales_Amount"] > 10000

print(high_sales_Amount)
```

```python
expensive_products = df[df["Unit_Price"] > 5000]

print(expensive_products)
```

```python
electronics_products = df[df["Category"] == "Electronics"]

print(electronics_products)
```

```python
samsung_products = df[df["Brand"] == "Samsung"]

print(samsung_products)
```

```python
multiple_quantity = df[df["Quantity"] > 1]

print(multiple_quantity)
```

```python
filtered_data = df[
    (df["Sales_Amount"] > 10000) &
    (df["Category"] == "Electronics")
]

print(filtered_data)
```

visualization:

```python
plt.figure(figsize=(8,5))
plt.hist(df["Sales_Amount"], bins=10)
plt.title("Sales Distribution")
plt.xlabel("Sales")
plt.ylabel("Frequency")
plt.show()
```

```python
category_sales = df.groupby("Category")["Sales_Amount"].sum()

plt.figure(figsize=(8,5))
category_sales.plot(kind="bar")
plt.title("Category-wise Sales")
plt.xlabel("Category")
plt.ylabel("Total Sales")
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.scatter(df["Unit_Price"], df["Sales_Amount"])
plt.title("Price vs Sales")
plt.xlabel("Price")
plt.ylabel("Sales")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(y=df["Sales_Amount"])
plt.title("Sales Box Plot")
plt.show()
```

```python
plt.figure(figsize=(8,5))

numeric_df = df.select_dtypes(include=['number'])

sns.heatmap(
    numeric_df.corr(),
    annot=True,
    cmap="coolwarm"
)

plt.title("Correlation Heatmap")
plt.show()
```

```python
category_count = df["Category"].value_counts()

plt.figure(figsize=(8,5))
plt.pie(
    category_count,
    labels=category_count.index,
    autopct="%1.1f%%"
)

plt.title("Category Distribution")
plt.show()
```
